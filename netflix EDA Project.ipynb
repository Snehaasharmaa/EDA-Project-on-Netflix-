```python
# exploring the distribution of content across different countries and regions
```

```python
# Fill missing values
df['country'] = df['country'].fillna('Unknown')

# Split multiple countries into separate rows
df_countries = df.assign(country=df['country'].str.split(', ')).explode('country')

# Inspect
print(df_countries[['title','country']].head())
```

```python
# Count titles per country
country_counts = df_countries['country'].value_counts().reset_index()
country_counts.columns = ['country', 'count']

# Top 15 countries
top_countries = country_counts.head(15)
print(top_countries)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='count', y='country', data=top_countries, palette='viridis')
plt.title('Top 15 Countries by Netflix Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# Count titles per country per year
country_year_trends = df_countries.groupby(['release_year','country']).size().reset_index(name='count')

# Top 5 countries for trend visualization
top_5_countries = top_countries['country'].head(5)

plt.figure(figsize=(12,6))
sns.lineplot(data=country_year_trends[country_year_trends['country'].isin(top_5_countries)],
             x='release_year', y='count', hue='country', marker='o')
plt.title('Netflix Content Trends Over Time for Top 5 Countries')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(title='Country')
plt.show()
```

# insights we got:

1) Top content-producing countries: Usually US, India, UK dominate.

2) Emerging regions: Growth in countries like Spain, South Korea, or Japan over recent years.

3) Trends over time: Some countries may see sharp increases in content production, showing Netflix‚Äôs regional expansion.

4) Diversity of content: Helps identify whether Netflix is producing more local content or global content

### Correlation Analysis:
### Investigating potential correlations between variables.


```python
df['duration'] = df['duration'].fillna('0 min')

# Create numeric columns
df['movie_duration'] = df.apply(lambda x: int(x['duration'].split()[0]) if x['type']=='Movie' and 'min' in x['duration'] else None, axis=1)
df['tv_seasons'] = df.apply(lambda x: int(x['duration'].split()[0]) if x['type']=='TV Show' and 'Season' in x['duration'] else None, axis=1)
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Correlation for Movies (without rating column)
movies = df[df['type']=='Movie'][['movie_duration','release_year']].dropna()
print("Movies Correlation:\n", movies.corr())

sns.heatmap(movies.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Movies')
plt.show()
```

```python
tv = df[df['type']=='TV Show'][['tv_seasons','release_year']].dropna()
print("TV Shows Correlation:\n", tv.corr())

sns.heatmap(tv.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for TV Shows')
plt.show()
```

## Content Evolution Over Time

```python
# Exploring how the characteristics of content (e.g., duration, ratings) have evolved over th eyears
```

```python
# Fill missing durations
df['duration'] = df['duration'].fillna('0 min')

# Numeric columns
df['movie_duration'] = df.apply(lambda x: int(x['duration'].split()[0]) 
                                if x['type']=='Movie' and 'min' in x['duration'] else None, axis=1)
df['tv_seasons'] = df.apply(lambda x: int(x['duration'].split()[0]) 
                            if x['type']=='TV Show' and 'Season' in x['duration'] else None, axis=1)
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
# Group by release year and get average movie duration
movies_trend = df[df['type']=='Movie'].groupby('release_year')['movie_duration'].mean().dropna()

plt.figure(figsize=(12,6))
sns.lineplot(x=movies_trend.index, y=movies_trend.values, marker='o', color='blue')
plt.title('Average Movie Duration Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (minutes)')
plt.show()
```

## insight:

Shows whether Netflix movies are becoming longer, shorter, or staying consistent over time.

Typically, Netflix movie duration remains fairly stable, around 90‚Äì120 minutes.

```python
# Group by release year and get average number of seasons
tv_trend = df[df['type']=='TV Show'].groupby('release_year')['tv_seasons'].mean().dropna()

plt.figure(figsize=(12,6))
sns.lineplot(x=tv_trend.index, y=tv_trend.values, marker='o', color='orange')
plt.title('Average Number of TV Show Seasons Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Seasons')
plt.show()
```

## Insight:

Older shows generally have more seasons, while newer shows often start with 1‚Äì2 seasons.

Indicates that long-running series accumulate seasons over time, while recent content is mostly new releases.

```python
# Analyzing content ratings over the years
```

```python
# Fill missing ratings
df['rating'] = df['rating'].fillna('Unknown')

# Count of ratings per year
ratings_trend = df.groupby(['release_year','rating']).size().reset_index(name='count')

plt.figure(figsize=(12,6))
sns.lineplot(data=ratings_trend, x='release_year', y='count', hue='rating')
plt.title('Distribution of Content Ratings Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Insight:

Tracks how Netflix‚Äôs content ratings have evolved.

For example, an increase in TV-MA or R titles over the years may indicate a shift toward more adult content.

Shows how the platform targets different age groups over time.

## Summary Insight for this Analysis

Movie durations have remained fairly consistent over time.

TV show seasons are higher for older shows; newer shows usually have fewer seasons.

Content ratings distribution shows trends in age targeting, reflecting Netflix‚Äôs evolving strategy to cater to both family and adult audiences.

#  üìå  Conclusions and Recommendations:

### 1. Content Characteristics and Trends

Movie Duration: Remains fairly consistent over the years (mostly 90‚Äì120 minutes). No strong trend toward longer or shorter films.

TV Show Seasons: Older shows tend to have more seasons; newer releases usually start with 1‚Äì2 seasons.

Content Ratings: Distribution shows Netflix produces content for all age groups, with an increase in adult-oriented content (TV-MA, R) over time.

#### Recommendation:

Maintain a balance of short and long-form content to cater to both casual viewers and binge-watchers.

Continue producing limited-series or short-season shows to attract new audiences quickly.

Monitor rating trends to ensure content aligns with target demographics.

### 2. User Preferences (Inferred from Content Distribution)

Popular Types: Netflix has more movies than TV shows, but some genres (Reality, Docuseries) are TV-show heavy.

Popular Genres: Drama, Comedy, and Documentary dominate the catalog, indicating audience preference.

Type-Genre Combinations: Action and Thriller are mostly movies; Reality and Docuseries are mostly TV shows.

#### Recommendation:

Focus on producing more content in high-demand genres (Drama, Comedy, Documentary) while exploring niche genres to attract diverse audiences.

Align content type (Movie vs TV Show) with genre-based preferences to maximize engagement.

### 3. Content Variety

Netflix offers a wide range of genres and categories, showing strong content diversity.

Regional content is concentrated in the US, followed by India, UK, and Japan.

#### Recommendation:

Expand production in underrepresented countries to capture regional audiences.

Continue offering diverse genres to satisfy global audience tastes.
